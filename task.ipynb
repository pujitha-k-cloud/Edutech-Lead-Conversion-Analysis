```python
# Step 3: Clean column names and multiline values

# 1ï¸âƒ£ Clean the column names (remove everything after "\n")
df.columns = [col.split('\n')[0].strip() for col in df.columns]

# 2ï¸âƒ£ Clean each cell: if there's a newline, keep only the first value before it
df = df.applymap(lambda x: str(x).split('\n')[0].strip() if isinstance(x, str) else x)

# 3ï¸âƒ£ Quick check â€” print cleaned columns and sample values
print("=== Cleaned Column Names ===")
print(df.columns.tolist())

print("\n=== Sample values after cleaning ===")
print(df.head(3))
```

```python
df.info()
```

```python
# Step 4: Convert dates and numeric fields

# Convert date columns to datetime format
date_columns = ['Date of Lead Generation', 'Date of Last Contact', 'Conversion Date']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce', dayfirst=True)

# Convert numeric fields to numbers (coerce errors to NaN)
num_columns = ['Sales Amount', 'Lead Score']
for col in num_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check results
print("\n=== Data types after conversion ===")
print(df.dtypes)

print("\n=== Sample data after type conversion ===")
print(df.head(3))
```

```python
# Step 5: Handle missing values and duplicates

# 1ï¸âƒ£ Count how many nulls exist before cleaning
print("=== Missing Values per Column (Before Cleaning) ===")
print(df.isnull().sum())

# 2ï¸âƒ£ Fill numeric NaN with median (robust against outliers)
for col in df.select_dtypes(include=['number']):
    df[col].fillna(df[col].median(), inplace=True)

# 3ï¸âƒ£ Fill categorical NaN with 'Unknown'
for col in df.select_dtypes(include=['object']):
    df[col].fillna('Unknown', inplace=True)

# 4ï¸âƒ£ Remove duplicate rows based on Lead ID (most unique field)
initial_rows = df.shape[0]
df.drop_duplicates(subset=['Lead ID'], inplace=True)
print(f"\nRemoved {initial_rows - df.shape[0]} duplicate rows (if any).")

# 5ï¸âƒ£ Verify everything is clean
print("\n=== Missing Values per Column (After Cleaning) ===")
print(df.isnull().sum())

print("\n=== Data Snapshot ===")
print(df.head(3))
```

```python
# Step 6: Standardize text and validate logical consistency

# 1ï¸âƒ£ Standardize text columns (consistent casing, trim spaces)
def clean_text(x):
    if isinstance(x, str):
        return x.strip().title()
    return x

text_columns = ['Lead Source', 'Lead Status', 'Region', 'Salesperson', 
                'Lead Qualification', 'Course Type', 'Follow-up Status', 
                'Marketing Campaign', 'Student Type']
for col in text_columns:
    if col in df.columns:
        df[col] = df[col].apply(clean_text)

# 2ï¸âƒ£ Check for logical date inconsistencies
invalid_dates = df[df['Conversion Date'] < df['Date of Lead Generation']]
print(f"Invalid date rows found: {invalid_dates.shape[0]}")

# If conversion happened before lead generation, drop or fix them
df.loc[df['Conversion Date'] < df['Date of Lead Generation'], 'Conversion Date'] = pd.NaT

# 3ï¸âƒ£ Preview cleaned data
print("\n=== Cleaned & Standardized Data Sample ===")
print(df.head(5))
```

```python
# Step 7: Post-cleaning validation checks

print("=== FINAL SHAPE OF DATA ===")
print(df.shape)

print("\n=== Remaining Missing Values ===")
print(df.isnull().sum())

print("\n=== Duplicate Count ===")
print(df.duplicated(subset=['Lead ID']).sum())

# Quick sanity check: any conversion before last contact? (shouldn't happen)
invalid_conversion = df[df['Conversion Date'] < df['Date of Last Contact']]
print(f"\nLeads with conversion before last contact: {invalid_conversion.shape[0]}")

# Sample final preview
print("\n=== Final Cleaned Data Sample ===")
print(df.head(5))
```

```python
# Step 8: Conversion Rate by Region

# 1ï¸âƒ£ Create a flag for converted leads
df['Is_Converted'] = df['Lead Status'].apply(lambda x: 1 if 'Converted' in str(x) else 0)

# 2ï¸âƒ£ Group by region and calculate metrics
region_conv = df.groupby('Region').agg(
    Total_Leads=('Lead ID', 'count'),
    Converted_Leads=('Is_Converted', 'sum')
)
region_conv['Conversion_Rate (%)'] = (region_conv['Converted_Leads'] / region_conv['Total_Leads'] * 100).round(2)

# 3ï¸âƒ£ Display summary
print("\n=== Conversion Rate by Region ===")
print(region_conv)

# 4ï¸âƒ£ Plot
import matplotlib.pyplot as plt
region_conv['Conversion_Rate (%)'].plot(kind='bar', figsize=(8,5))
plt.title("Conversion Rate by Region")
plt.ylabel("Conversion Rate (%)")
plt.xlabel("Region")
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 9: Sales Performance by Salesperson

sales_perf = df.groupby('Salesperson').agg(
    Total_Leads=('Lead ID', 'count'),
    Converted_Leads=('Is_Converted', 'sum'),
    Total_Sales=('Sales Amount', 'sum'),
    Average_Sales=('Sales Amount', 'mean')
).round(2)

sales_perf['Conversion_Rate (%)'] = (sales_perf['Converted_Leads'] / sales_perf['Total_Leads'] * 100).round(2)

print("\n=== Sales Performance by Salesperson ===")
print(sales_perf)

# Plot total sales and conversion rate
fig, ax1 = plt.subplots(figsize=(8,5))

sales_perf['Total_Sales'].plot(kind='bar', ax=ax1, color='skyblue', position=0, width=0.4)
ax1.set_ylabel('Total Sales')
ax1.set_title('Sales Performance by Salesperson')

ax2 = ax1.twinx()
sales_perf['Conversion_Rate (%)'].plot(kind='line', ax=ax2, color='red', marker='o')
ax2.set_ylabel('Conversion Rate (%)')

plt.show()
```

```python
# Step 10: Lead Source vs Conversion Rate

lead_source_conv = df.groupby('Lead Source').agg(
    Total_Leads=('Lead ID', 'count'),
    Converted_Leads=('Is_Converted', 'sum')
)
lead_source_conv['Conversion_Rate (%)'] = (lead_source_conv['Converted_Leads'] / lead_source_conv['Total_Leads'] * 100).round(2)

print("\n=== Lead Source vs Conversion Rate ===")
print(lead_source_conv)

# Plot conversion rates
lead_source_conv['Conversion_Rate (%)'].sort_values(ascending=False).plot(
    kind='bar', figsize=(8,5), color='orange'
)
plt.title("Conversion Rate by Lead Source")
plt.ylabel("Conversion Rate (%)")
plt.xlabel("Lead Source")
plt.xticks(rotation=45)
plt.show()
```
